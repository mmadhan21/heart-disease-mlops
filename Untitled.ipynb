```python
import subprocess
import os

print("=" * 55)
print("  K8S DEPLOYMENT EVIDENCE")
print("=" * 55)

# Fix PATH first
extra_paths = [
    r"C:\Program Files\Docker\Docker\resources\bin",
    r"C:\ProgramData\DockerDesktop\version-bin",
    r"C:\Windows\System32",
]
for p in extra_paths:
    if os.path.exists(p):
        os.environ["PATH"] = (
            p + ";" + os.environ["PATH"]
        )

# Run all kubectl commands with shell=True
commands = [
    (
        "kubectl get pods "
        "-l app=heart-disease-api -o wide",
        "Pods Status"
    ),
    (
        "kubectl get service "
        "heart-disease-service",
        "Service"
    ),
    (
        "kubectl get deployment "
        "heart-disease-api",
        "Deployment"
    ),
    (
        "kubectl get all "
        "-l app=heart-disease-api",
        "All Resources"
    ),
]

for cmd, label in commands:
    print(f"\n  {label}:")
    print("  " + "-" * 50)
    r = subprocess.run(
        cmd,
        shell=True,
        capture_output=True,
        text=True
    )
    if r.returncode == 0:
        for line in r.stdout.strip().split("\n"):
            print(f"  {line}")
    else:
        print(f"  Error: {r.stderr[:100]}")

print()
print("=" * 55)
print("=" * 55)
```
