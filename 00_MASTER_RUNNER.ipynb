#Model Comparison

```python
import joblib
import pandas as pd
import sys
import os

os.environ["GIT_PYTHON_REFRESH"] = "quiet"
sys.path.insert(0, "src")

from preprocess import (
    load_data, clean_data,
    get_features_and_target, split_data
)
from sklearn.metrics import (
    accuracy_score, precision_score,
    recall_score, f1_score, roc_auc_score
)

# Reload data with same split
df = load_data("data/heart_cleaned.csv")
df = clean_data(df)
X, y = get_features_and_target(df)
X_train, X_test, y_train, y_test = split_data(X, y)

# Compare both models
results = {}
for model_name, path in [
    ("Logistic Regression", "models/Logistic_Regression.pkl"),
    ("Random Forest",       "models/Random_Forest.pkl"),
]:
    m      = joblib.load(path)
    y_pred = m.predict(X_test)
    y_prob = m.predict_proba(X_test)[:, 1]

    results[model_name] = {
        "Accuracy"  : round(accuracy_score (y_test, y_pred), 4),
        "Precision" : round(precision_score(y_test, y_pred), 4),
        "Recall"    : round(recall_score   (y_test, y_pred), 4),
        "F1 Score"  : round(f1_score       (y_test, y_pred), 4),
        "ROC-AUC"   : round(roc_auc_score  (y_test, y_prob), 4),
    }

# Display comparison table
results_df = pd.DataFrame(results).T
print("=" * 55)
print("  FINAL MODEL COMPARISON TABLE")
print("=" * 55)
display(results_df)

# Best model
best = results_df["ROC-AUC"].idxmax()
print(f"\n  🏆  Best Model : {best}")
print(f"      ROC-AUC    : {results_df.loc[best, 'ROC-AUC']}")
print(f"      Accuracy   : {results_df.loc[best, 'Accuracy']}")
print(f"      Recall     : {results_df.loc[best, 'Recall']}")
print("=" * 55)
```

#Start Docker Container

```python
import subprocess
import socket
import time

print("=" * 55)
print("  STARTING DOCKER CONTAINER")
print("=" * 55)

# Clean up any old container
subprocess.run(["docker", "stop", "heart-api"],
               capture_output=True)
subprocess.run(["docker", "rm",   "heart-api"],
               capture_output=True)
time.sleep(2)
print("  Cleaned up old containers")

# Start fresh container
result = subprocess.run(
    [
        "docker", "run",
        "--detach",
        "--publish", "8000:8000",
        "--name", "heart-api",
        "--env",
        "MODEL_PATH=/app/models/best_model.pkl",
        "--env", "PYTHONUNBUFFERED=1",
        "heart-disease-api:v1"
    ],
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print(f"  Container started!")
    print(f"  ID : {result.stdout.strip()[:12]}...")
    print(f"  Name : heart-api")
    print(f"  Port : 8000 -> 8000")
else:
    print(f"  Failed to start!")
    print(f"  Error: {result.stderr}")

# Watch live logs while waiting
print("\n  Live logs while initializing:\n")
print("-" * 55)

ready = False
for i in range(10):
    time.sleep(5)
    elapsed = (i + 1) * 5

    # Get latest logs
    logs = subprocess.run(
        ["docker", "logs", "heart-api",
         "--tail", "4"],
        capture_output=True, text=True
    )
    log_text = (
        logs.stdout + logs.stderr
    ).strip()

    print(f"  [{elapsed:>2}s] {log_text[:80]}")

    # Check for success
    if "startup complete" in log_text.lower():
        print(f"\n  App started successfully!")
        ready = True
        break

    # Check for errors
    if "error" in log_text.lower():
        print(f"\n  Error in logs!")
        full_logs = subprocess.run(
            ["docker", "logs", "heart-api"],
            capture_output=True, text=True
        )
        print(full_logs.stdout)
        print(full_logs.stderr)
        break

    # Check port
    with socket.socket(
        socket.AF_INET, socket.SOCK_STREAM
    ) as s:
        s.settimeout(2)
        if s.connect_ex(("127.0.0.1", 8000)) == 0:
            print(f"\n  Port 8000 OPEN after {elapsed}s!")
            ready = True
            break

print("-" * 55)
if ready:
    print("\n  Container is READY!")
    print("  API available at: http://localhost:8000")
else:
    print("\n  Still starting — run next cell")

print("=" * 55)
```

#Test Running Container

```python
import urllib.request
import json
import socket

print("=" * 55)
print("  API ENDPOINT TESTING")
print("=" * 55)

# Helper function
def test_get(url, label):
    print(f"\n  {label}")
    print(f"  GET {url}")
    try:
        with urllib.request.urlopen(
            url, timeout=10
        ) as r:
            data = json.loads(r.read())
            print(f"  Status : 200 OK")
            print(f"  Response: {data}")
            return True
    except Exception as e:
        print(f"  FAILED: {e}")
        return False

def test_post(url, payload, label):
    print(f"\n  {label}")
    print(f"  POST {url}")
    try:
        data = json.dumps(payload).encode("utf-8")
        req  = urllib.request.Request(
            url,
            data    = data,
            headers = {
                "Content-Type": "application/json"
            },
            method = "POST"
        )
        with urllib.request.urlopen(
            req, timeout=10
        ) as r:
            result = json.loads(r.read())
            print(f"  Status    : 200 OK")
            print(f"  Prediction: {result['prediction']}")
            print(f"  Diagnosis : {result['diagnosis']}")
            print(f"  Confidence: {result['confidence']}")
            print(f"  Risk Level: {result['risk_level']}")
            print(f"  Latency   : {result['latency_ms']}ms")
            return True
    except Exception as e:
        print(f"  FAILED: {e}")
        return False

# Run all tests
base = "http://localhost:8000"
results = {}

results["root"] = test_get(
    f"{base}/",
    "TEST 1: Root endpoint"
)

results["health"] = test_get(
    f"{base}/health",
    "TEST 2: Health endpoint"
)

results["predict"] = test_post(
    f"{base}/predict",
    {
        "age": 63, "sex": 1, "cp": 3,
        "trestbps": 145, "chol": 233,
        "fbs": 1, "restecg": 0,
        "thalach": 150, "exang": 0,
        "oldpeak": 2.3, "slope": 0,
        "ca": 0, "thal": 1
    },
    "TEST 3: Predict endpoint (High Risk Patient)"
)

results["predict2"] = test_post(
    f"{base}/predict",
    {
        "age": 35, "sex": 0, "cp": 1,
        "trestbps": 110, "chol": 180,
        "fbs": 0, "restecg": 0,
        "thalach": 185, "exang": 0,
        "oldpeak": 0.0, "slope": 2,
        "ca": 0, "thal": 2
    },
    "TEST 4: Predict endpoint (Low Risk Patient)"
)

# Test invalid input
print(f"\n  TEST 5: Invalid input (expect 422)")
try:
    data = json.dumps(
        {"age": 63}
    ).encode("utf-8")
    req  = urllib.request.Request(
        f"{base}/predict",
        data    = data,
        headers = {
            "Content-Type": "application/json"
        },
        method  = "POST"
    )
    urllib.request.urlopen(req, timeout=10)
    print("  FAILED: Should have returned 422")
    results["invalid"] = False
except urllib.error.HTTPError as e:
    if e.code == 422:
        print(f"  Status : 422 Unprocessable Entity")
        print(f"  Correctly rejected invalid input!")
        results["invalid"] = True
    else:
        print(f"  Got HTTP {e.code}")
        results["invalid"] = False

# Summary
print()
print("=" * 55)
print("  TEST SUMMARY")
print("=" * 55)
passed = 0
for test, result in results.items():
    icon = "PASS" if result else "FAIL"
    print(f"  [{icon}]  {test}")
    if result:
        passed += 1

print()
print(f"  {passed}/{len(results)} tests passed")
print()
print("=" * 55)
```

```python
import subprocess
import os
import time

print("=" * 55)
print("  DEPLOYING TO KUBERNETES")
print("=" * 55)

# Find Docker
r = subprocess.run(
    "where docker", shell=True,
    capture_output=True, text=True
)
docker_dir = os.path.dirname(
    r.stdout.strip().split("\n")[0].strip()
)
os.environ["PATH"] = (
    docker_dir + ";" + os.environ["PATH"]
)

# Clean old deployment
print("\n  Cleaning old deployment...")
subprocess.run(
    ["kubectl", "delete", "deployment",
     "heart-disease-api",
     "--ignore-not-found=true"],
    capture_output=True
)
subprocess.run(
    ["kubectl", "delete", "service",
     "heart-disease-service",
     "--ignore-not-found=true"],
    capture_output=True
)
time.sleep(5)
print("  Cleaned!")

# Write deployment manifest
deployment_yaml = """apiVersion: apps/v1
kind: Deployment
metadata:
  name: heart-disease-api
  namespace: default
  labels:
    app: heart-disease-api
    version: v1
spec:
  replicas: 1
  selector:
    matchLabels:
      app: heart-disease-api
  template:
    metadata:
      labels:
        app: heart-disease-api
    spec:
      containers:
        - name: heart-disease-api
          image: heart-disease-api:v1
          imagePullPolicy: Never
          ports:
            - containerPort: 8000
          env:
            - name: MODEL_PATH
              value: "/app/models/best_model.pkl"
            - name: PYTHONUNBUFFERED
              value: "1"
          resources:
            requests:
              memory: "256Mi"
              cpu: "250m"
            limits:
              memory: "512Mi"
              cpu: "500m"
          readinessProbe:
            httpGet:
              path: /health
              port: 8000
            initialDelaySeconds: 30
            periodSeconds: 10
            failureThreshold: 5
          livenessProbe:
            httpGet:
              path: /health
              port: 8000
            initialDelaySeconds: 45
            periodSeconds: 15
            failureThreshold: 5
"""

service_yaml = """apiVersion: v1
kind: Service
metadata:
  name: heart-disease-service
  namespace: default
spec:
  type: NodePort
  selector:
    app: heart-disease-api
  ports:
    - protocol: TCP
      port: 80
      targetPort: 8000
      nodePort: 30080
"""

# Save manifests
with open("deployment/deployment.yaml",
          "w", encoding="utf-8") as f:
    f.write(deployment_yaml)

with open("deployment/service.yaml",
          "w", encoding="utf-8") as f:
    f.write(service_yaml)

print("\n  Manifests saved!")

# Apply
print("\n  Applying deployment...")
r1 = subprocess.run(
    ["kubectl", "apply",
     "-f", "deployment/deployment.yaml"],
    capture_output=True, text=True
)
print(f"  {r1.stdout.strip()}")
if r1.stderr:
    print(f"  Note: {r1.stderr.strip()[:100]}")

print("\n  Applying service...")
r2 = subprocess.run(
    ["kubectl", "apply",
     "-f", "deployment/service.yaml"],
    capture_output=True, text=True
)
print(f"  {r2.stdout.strip()}")

# Watch pods
print("\n  Watching pods (2 min max)...")
print("-" * 55)

success = False
for i in range(12):
    time.sleep(10)
    elapsed = (i + 1) * 10

    r = subprocess.run(
        ["kubectl", "get", "pods",
         "-l", "app=heart-disease-api"],
        capture_output=True, text=True
    )
    status = r.stdout.strip()
    print(f"\n  [{elapsed}s]")
    print(f"  {status}")

    # Success
    if "Running" in status:
        print("\n  Pod is RUNNING!")
        success = True

        # Get pod name
        pod_r = subprocess.run(
            ["kubectl", "get", "pods",
             "-l", "app=heart-disease-api",
             "-o", "name"],
            capture_output=True, text=True
        )
        pod = pod_r.stdout.strip().split("\n")[0]

        # Show logs
        logs = subprocess.run(
            ["kubectl", "logs",
             pod, "--tail", "10"],
            capture_output=True, text=True
        )
        print("\n  Pod logs:")
        print(logs.stdout)
        break

    # Still pulling
    if "Pending" in status or \
       "ContainerCreating" in status:
        print("  Still starting... keep waiting")
        continue

    # Image error
    if "ErrImage" in status:
        print("\n  Image error still!")
        pod_r = subprocess.run(
            ["kubectl", "get", "pods",
             "-l", "app=heart-disease-api",
             "-o", "name"],
            capture_output=True, text=True
        )
        pod = pod_r.stdout.strip().split("\n")[0]
        desc = subprocess.run(
            ["kubectl", "describe", pod],
            capture_output=True, text=True
        )
        if "Events:" in desc.stdout:
            print(desc.stdout[
                desc.stdout.find("Events:"):][:500]
            )
        break

    # Crash
    if "CrashLoop" in status:
        print("\n  CrashLoop!")
        pod_r = subprocess.run(
            ["kubectl", "get", "pods",
             "-l", "app=heart-disease-api",
             "-o", "name"],
            capture_output=True, text=True
        )
        pod = pod_r.stdout.strip().split("\n")[0]
        logs = subprocess.run(
            ["kubectl", "logs",
             pod, "--tail", "20"],
            capture_output=True, text=True
        )
        print(logs.stdout)
        print(logs.stderr)
        break

print("=" * 55)
if success:
    print("  DEPLOYMENT SUCCESSFUL!")
else:
    print("  Deployment needs attention")
    print("  Paste output above!")
print("=" * 55)
```

```python
import urllib.request
import json
import time
import random

print("=" * 55)
print("  GENERATING API TRAFFIC")
print("  For Prometheus metrics")
print("=" * 55)

# Sample patients
patients = [
    {
        "age": 63, "sex": 1, "cp": 3,
        "trestbps": 145, "chol": 233,
        "fbs": 1, "restecg": 0,
        "thalach": 150, "exang": 0,
        "oldpeak": 2.3, "slope": 0,
        "ca": 0, "thal": 1
    },
    {
        "age": 35, "sex": 0, "cp": 1,
        "trestbps": 110, "chol": 180,
        "fbs": 0, "restecg": 0,
        "thalach": 185, "exang": 0,
        "oldpeak": 0.0, "slope": 2,
        "ca": 0, "thal": 2
    },
    {
        "age": 55, "sex": 1, "cp": 2,
        "trestbps": 130, "chol": 250,
        "fbs": 0, "restecg": 1,
        "thalach": 160, "exang": 0,
        "oldpeak": 1.5, "slope": 1,
        "ca": 1, "thal": 2
    },
    {
        "age": 45, "sex": 0, "cp": 0,
        "trestbps": 120, "chol": 200,
        "fbs": 0, "restecg": 0,
        "thalach": 170, "exang": 0,
        "oldpeak": 0.5, "slope": 2,
        "ca": 0, "thal": 2
    },
]

# Send 20 requests
print("\n  Sending 20 requests to API...")
print("-" * 55)

success = 0
errors  = 0

for i in range(20):
    patient = patients[i % len(patients)]

    try:
        payload = json.dumps(patient).encode()
        req = urllib.request.Request(
            "http://localhost:8000/predict",
            data    = payload,
            headers = {
                "Content-Type": "application/json"
            },
            method  = "POST"
        )
        with urllib.request.urlopen(
            req, timeout=5
        ) as resp:
            result = json.loads(resp.read())
            success += 1
            print(
                f"  [{i+1:>2}] "
                f"Pred={result['prediction']} | "
                f"Risk={result['risk_level']:<6} | "
                f"{result['latency_ms']}ms"
            )
    except Exception as e:
        errors += 1
        print(f"  [{i+1:>2}] Error: {str(e)[:40]}")

    time.sleep(0.5)

print("-" * 55)
print(f"\n  Total   : 20 requests")
print(f"  Success : {success}")
print(f"  Errors  : {errors}")
print()
print("  Metrics now visible in Prometheus!")
print("=" * 55)
```

```python
import urllib.request
import json
import time

print("=" * 55)
print("  PROMETHEUS VERIFICATION")
print("=" * 55)

time.sleep(5)

# Check Prometheus targets
print("\n  Checking Prometheus targets...")
try:
    with urllib.request.urlopen(
        "http://localhost:9090/api/v1/targets",
        timeout=10
    ) as resp:
        data = json.loads(resp.read())
        targets = data.get(
            "data", {}
        ).get("activeTargets", [])

        print(f"  Active targets: {len(targets)}")
        for t in targets:
            job    = t.get("labels", {}).get("job")
            health = t.get("health")
            url    = t.get(
                "scrapeUrl", ""
            )[:50]
            icon   = "OK" if health=="up" else "DOWN"
            print(f"  [{icon}]  {job:<25} {health}")
            print(f"        {url}")

except Exception as e:
    print(f"  Error: {e}")
    print("  Prometheus may still be starting")
    print("  Wait 30s and try again")

# Check our custom metrics
print("\n  Checking API metrics...")
try:
    with urllib.request.urlopen(
        "http://localhost:9090/api/v1/query"
        "?query=heart_api_requests_total",
        timeout=10
    ) as resp:
        data = json.loads(resp.read())
        results = data.get(
            "data", {}
        ).get("result", [])

        if results:
            val = results[0].get(
                "value", [0, 0]
            )[1]
            print(f"  Total requests: {val}")
            print("  Prometheus IS scraping API!")
        else:
            print("  Metric not found yet")
            print("  Wait 15s and refresh")

except Exception as e:
    print(f"  Query error: {e}")

print("=" * 55)
```

#Push to GIT

```python
import subprocess
import os

print("=" * 55)
print("  INITIALIZE GIT REPOSITORY")
print("=" * 55)

GITHUB_USERNAME = "mmadhan21"
REPO_NAME       = "heart-disease-mlops"

REPO_URL = (
    f"https://github.com/"
    f"{GITHUB_USERNAME}/{REPO_NAME}.git"
)
print(f"  Repo URL: {REPO_URL}")

# Step 1: Initialize git
print("\n  Step 1: Initializing git...")
r1 = subprocess.run(
    "git init",
    shell=True,
    capture_output=True,
    text=True,
    cwd=os.getcwd()
)
print(f"  {r1.stdout.strip()}")

# Step 2: Add remote
print("\n  Step 2: Adding remote...")
# Remove existing remote if any
subprocess.run(
    "git remote remove origin",
    shell=True,
    capture_output=True
)
r2 = subprocess.run(
    f"git remote add origin {REPO_URL}",
    shell=True,
    capture_output=True,
    text=True
)
if r2.returncode == 0:
    print(f"  Remote added: {REPO_URL}")
else:
    print(f"  Error: {r2.stderr[:100]}")

# Step 3: Create .gitignore additions
extra_gitignore = (
    "\n# Extra ignores\n"
    "heart-api.tar\n"
    "*.tar\n"
    "mlops_env/\n"
    "__pycache__/\n"
    "*.pyc\n"
    ".ipynb_checkpoints/\n"
    "api_requests.log\n"
)
with open(".gitignore", "a",
          encoding="utf-8") as f:
    f.write(extra_gitignore)
print("\n  Step 3: .gitignore updated!")

# Step 4: Stage all files
print("\n  Step 4: Staging files...")
r4 = subprocess.run(
    "git add .",
    shell=True,
    capture_output=True,
    text=True
)
if r4.returncode == 0:
    print("  All files staged!")
else:
    print(f"  Error: {r4.stderr[:100]}")

# Step 5: Check what will be committed
r5 = subprocess.run(
    "git status --short",
    shell=True,
    capture_output=True,
    text=True
)
lines = r5.stdout.strip().split("\n")
print(f"\n  Files to commit: {len(lines)}")
for line in lines[:15]:
    print(f"  {line}")
if len(lines) > 15:
    print(f"  ... and {len(lines)-15} more")

print("=" * 55)
```
